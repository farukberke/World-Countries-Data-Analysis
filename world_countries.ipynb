## World Countries Data Analysis

### Data & Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("countries.csv")
```

### Missing Values

```python
print("Empty Values: ", df.isna().sum().sum())
```

```python
df.isna().sum()
```

```python
df[df["Agriculture"].isnull()]["Country"]
```

```python
df[df["Agriculture"].isna()] = df[df["Agriculture"].isna()].fillna(0)
```

```python
df[df["Agriculture"].isnull()].sum()
```

```python
df[df["Literacy (%)"].isnull()]
```

```python
df.groupby("Region")["Literacy (%)"].mean().sort_values(ascending=False)
```

```python
df["Literacy (%)"] = df["Literacy (%)"].fillna(df.groupby("Region")["Literacy (%)"].transform("mean"))
```

```python
df.isna().sum()
```

```python
df["Climate"] = df["Climate"].fillna(df.groupby("Region")["Climate"].transform("mean"))
```

```python
df.isna().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df[["Other (%)","Crops (%)"]] = df[["Other (%)","Crops (%)"]].fillna(df.groupby("Region")[["Other (%)","Crops (%)"]].transform("mean"))
```

```python
df.isna().sum().sum()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20,20))
plt.subplots_adjust(hspace=0.35)
#Shape of subplots

corr_to_gdp = pd.Series() #corelation of columns to GDP
for col in df.columns.values[2:]:
    if((col != "GDP ($ per capita)")  &  (col != "Climate")):
        corr_to_gdp[col] = df["GDP ($ per capita)"].corr(df[col])
        

abs_corr_to_gdp = corr_to_gdp.abs().sort_values(ascending=False)


corr_to_gdp = corr_to_gdp.loc[abs_corr_to_gdp.index]


for i in range(3): # This is the plot
    for j in range(3):
        sns.regplot(x=corr_to_gdp.index.values[i*3+j], y="GDP ($ per capita)", data=df, ax=axes[i,j], fit_reg=False, marker="o")

        title = "Correlation : " + str(corr_to_gdp[i*3+j])
        axes[i,j].set_title(title)
```
